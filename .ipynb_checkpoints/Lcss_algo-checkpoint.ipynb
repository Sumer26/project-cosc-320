```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import time
sys.setrecursionlimit(10**6)
```

LCSS algorithm

```python
def LCSAlgorithm(X, Y, m, n):
    L = [[None] * (n + 1) for i in range(m + 1)]

    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0 or j == 0:
                L[i][j] = 0
            elif X[i - 1] == Y[j - 1]:
                L[i][j] = L[i - 1][j - 1] + 1
            else:
                L[i][j] = max(L[i - 1][j], L[i][j - 1])
    index = L[m][n]
    lcs = [""] * (index+1)
    lcs[index] = ""
    i = m
    j = n
    while i > 0 and j > 0:
        if X[i-1] == Y[j-1]:
            lcs[index-1] = X[i-1]
            i -= 1
            j -= 1
            index -= 1
        elif L[i-1][j] > L[i][j-1]:
            i -= 1
        else:
            j -= 1
    return "".join(lcs)
def dot_plot(X, Y):
    m = len(X)
    n = len(Y)
    plot = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            if X[i] == Y[j]:
                plot[i,j] = 1
    return plot
def LCSOutput(x, y, ax=None):
    len1 = len(x)
    len2 = len(y)
    lcs = LCSAlgorithm(x, y, len1, len2)
    print("LCS: " + lcs)
    if ax:
        plot = dot_plot(x, y)
        ax.imshow(plot, cmap='Greys', interpolation='nearest')
        ax.set_xticks(np.arange(-.5, len(y), 1))
        ax.set_yticks(np.arange(-.5, len(x), 1))
        ax.set_xticklabels([''] + list(y))
        ax.set_yticklabels([''] + list(x))
        ax.tick_params(axis=u'both', which=u'both',length=0)
    return lcs
def search_files(dir_path, search_word):
    t = []
    s = []
    for file_name in os.listdir(dir_path):
        if os.path.isdir(os.path.join(dir_path, file_name)) or not file_name.endswith(".txt"):
            continue
        file_path = os.path.join(dir_path, file_name)
        size = os.path.getsize(file_path)
        s.append(size)
        with open(file_path, 'r') as f:
            file_text = f.read()
        start_time = time.time()
        pattern_start = file_text.find(search_word)
        if pattern_start != -1:
            pattern_end = pattern_start + len(search_word)
            # Print the result for this file
            print(f"Pattern found in {file_path} at position {pattern_start}:")
            LCSOutput(search_word, file_text[pattern_start:pattern_end])
            print("\n")
        # Record the time taken
        end_time = time.time()
        elapsed_time = end_time - start_time
        t.append(elapsed_time)
    return s, t
```

```python
dir_path = "Files"
search_word = "apple"
sizes, times = search_files(dir_path, search_word)
plt.scatter(sizes, times)
plt.xlabel("File size")
plt.ylabel("Running time (seconds)")
n_values = np.linspace(min(sizes), max(sizes), 100)
plt.plot(n_values, (max(times) / max(sizes)**2) * n_values**2, label="O(n^2)")
plt.legend()
plt.show()
```
