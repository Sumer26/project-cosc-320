LCSS algorithm

```python

def LCSAlgorithm(a, b, x, y):
    r = [["" for j in range(y+1)] for i in range(x+1)]
    l = [[0 for j in range(y+1)] for i in range(x+1)]

    for i in range(1, x+1):
        l[i][0] = 0

    for j in range(1, y+1):
        l[0][j] = 0

    for i in range(1, x+1):
        for j in range(1, y+1):
            if a[i-1] == b[j-1]:
                l[i][j] = l[i-1][j-1] + 1
                r[i][j] = "↖"
            elif l[i-1][j] >= l[i][j-1]:
                l[i][j] = l[i-1][j]
                r[i][j] = "↑"
            else:
                l[i][j] = l[i][j-1]
                r[i][j] = "←"

    return l, r
def LCSOutput(r, a, i, j):
    if i == 0 or j == 0:
        return
    if r[i][j] == "↖":
        LCSOutput(r, a, i-1, j-1)
        print(a[i-1], end="")
    elif r[i][j] == "↑":
        LCSOutput(r, a, i-1, j)
    else:
        LCSOutput(r, a, i, j-1)
```

```python
# take search pattern as input
search_word = input("Enter the word pattern to search for: ")

# directory from which sample files are stored
dir_path = "Files"

# Iterateing over the files to search for a pattern
for file_name in os.listdir(dir_path):
    # Skips the directories or non-text files
    if os.path.isdir(os.path.join(dir_path, file_name)) or not file_name.endswith(".txt"):
        continue
    
    # Building the path
    file_path = os.path.join(dir_path, file_name)
    
    # Reading the text
    with open(file_path, 'r') as f:
        text = f.read()
        
    # Searching for pattern
    pattern_start = text.find(search_word)
    if pattern_start != -1:
        pattern_end = pattern_start + len(search_word)
        
        # Printing the result
        print(f"Pattern found in {file_path} at position {pattern_start}:")
        LCSOutput(*LCSAlgorithm(search_word, text, len(search_word), len(text)), len(search_word), len(text))
        print("\n")
```
